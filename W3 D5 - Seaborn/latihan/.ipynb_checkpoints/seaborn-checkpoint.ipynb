```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('flights')
df
```

# Relation chart

Line plot

```python
#matplotlib plt.plot()

plt.figure(figsize=(12,8))
sns.lineplot(x='year', y='passengers', data=df) #df['year]
plt.show()
```

```python
#matplotlib argumen color/c
plt.figure(figsize=(12,8))
sns.lineplot( x='year', y='passengers',marker='o', data=df, hue='month') #mengubah color saja gunakan palette, hue = color datapoint based on data
plt.show()
```

```python
data = pass_month.sort_values('passengers', ascending=False).head(5)

chart = sns.barplot(x = 'year', y='passengers',data=data)
# chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
for p in chart.patches:
    chart.annotate("{:,.2f}".format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```
